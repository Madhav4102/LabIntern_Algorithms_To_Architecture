```python
import os
import cv2
import numpy as np


frame_array_list=[]
```

```python
video_path= "test_video.mp4"
output_folder = "output_frames"



capture = cv2.VideoCapture(video_path)


width =0
height = 0


frame_count =0
while capture.isOpened():
    ret,frame = capture.read() # the frame is actually the numpy array itself
    
    if not ret:
        
        break
    
    if(frame_count ==0):
        width= frame.shape[1]
        height= frame.shape[0]
    
    
    r = (frame_count >> 16) & 255 
    g = (frame_count >> 8) & 255   
    b = frame_count & 255          

    extra_info_position = np.full((1, width, 3), [r, g, b], dtype=np.uint8)
    final_frame= np.vstack((frame,extra_info_position))
    frame_array_list.append(final_frame)


    frame_count += 1
    
    
capture.release() 

```

```python
import cv2
import numpy as np


sorted_frame_array_list = [None] * frame_count  


for frame in frame_array_list:
    sortingdata = frame[-1, 0, :]  
    extracted_frame = frame[:-1]  
    
    
    position = (sortingdata[0] << 16) + (sortingdata[1] << 8) + sortingdata[2]
    
    if position >= len(sorted_frame_array_list):  
        continue 
    
    sorted_frame_array_list[position] = extracted_frame  # Store in correct order


sorted_frame_array_list = [frame for frame in sorted_frame_array_list if frame is not None]


if not sorted_frame_array_list:
    raise ValueError("No valid frames found in sorted list.")


first_frame = sorted_frame_array_list[0]
height, width, _ = first_frame.shape
fps = 25

output_video = "output_video.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")  
out = cv2.VideoWriter(output_video, fourcc, fps, (width, height))


print(f"Total frames: {len(sorted_frame_array_list)}")  

for frame in sorted_frame_array_list:
    out.write(frame)  

out.release()
print(f"Video saved as {output_video}")
```
